```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
```

```python
and_data=pd.DataFrame()
or_data=pd.DataFrame()
xor_data=pd.DataFrame()
```

```python
and_data['input1']=[0,0,1,1]
and_data['input2']=[0,1,0,1]
and_data['output']=[0,0,0,1]
```

```python
and_data.head()
```

```python
sns.scatterplot(x='input1',y='input2',data=and_data,hue='output')
```

```python
X=and_data.iloc[:,0:2]
y=and_data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
p.fit(X,y)
```

```python
print(p.coef_)
print(p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values,y.values,clf=p,legend=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter points (your data)
plt.scatter(X['input1'], X['input2'], c=y, cmap='bwr', s=100)

# Decision boundary
x_vals = np.linspace(-1, 2, 100)
y_vals = -(p.coef_[0][0]*x_vals + p.intercept_[0]) / p.coef_[0][1]

plt.plot(x_vals, y_vals, 'k--', label='Decision Boundary')
 
```
