```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import requests
```

```python
data=pd.read_csv('placement.csv')
```

```python
data.head()
```

```python
sns.scatterplot(x='cgpa',y='resume_score',data=data,hue='placed')
```

```python
X=data.iloc[:,0:2]
y=data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
p= Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```
